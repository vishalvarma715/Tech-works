```python
# %%
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is already loaded

# Get numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns

# Create histograms for numeric features
for col in numeric_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col].dropna(), kde=True) # Drop NA values for plotting
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Create boxplots for numeric features
for col in numeric_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)
    plt.show()
```

```python
# %%
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is already loaded and numeric_cols are identified

# Create a pairplot for a subset of numeric features to visualize pairwise relationships
# Pairplot can be computationally intensive for many columns, so consider selecting key features.
# For demonstration, let's use 'Age', 'Fare', and 'Survived'.
# Ensure 'Survived' is treated as numeric for the pairplot
pairplot_cols = ['Age', 'Fare', 'Survived']

# Drop rows with any NaN values in the selected columns for pairplot clarity
df_pairplot = df[pairplot_cols].dropna()

# Create the pairplot
sns.pairplot(df_pairplot, diag_kind='kde') # Use kde for diagonal plots
plt.suptitle('Pairwise Relationships of Selected Numeric Features', y=1.02) # Add a title
plt.show()


# Create a correlation matrix heatmap for all numeric features
plt.figure(figsize=(12, 8))
correlation_matrix = df[numeric_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

```python
# 📌 Step 1: Install and Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set up plot style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# 📌 Step 2: Load Dataset
# If you upload via Colab interface:
# from google.colab import files
# uploaded = files.upload()

# Load Titanic dataset
df = pd.read_csv("Titanic-Dataset.csv")  # Make sure to upload the file manually

# 📌 Step 3: Basic Info
print("✅ Dataset Info:")
print(df.info())
print("\n🔍 First 5 Rows:")
print(df.head())

# 📌 Step 4: Survival Rate by Gender
survival_by_sex = df.groupby('Sex')['Survived'].mean()
plt.figure()
sns.barplot(x=survival_by_sex.index, y=survival_by_sex.values)
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.xlabel('Gender')
plt.show()

# 📌 Step 5: Survival Rate by Class
survival_by_class = df.groupby('Pclass')['Survived'].mean()
plt.figure()
sns.barplot(x=survival_by_class.index.astype(str), y=survival_by_class.values)
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.show()

# 📌 Step 6: Age Distribution with Survival Overlay
plt.figure()
sns.histplot(data=df, x='Age', hue='Survived', multiple='stack', bins=30)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# 📌 Step 7: Fare Boxplot (To Check for Outliers)
plt.figure()
sns.boxplot(x=df['Fare'])
plt.title('Fare Distribution (Outliers Visible)')
plt.xlabel('Fare')
plt.show()

# 📌 Step 8: Missing Values Summary
print("\n🚨 Missing Values:")
print(df.isnull().sum())
```
